```python
# Houman Alizadeh
# Homework Lecture 14
# Radioactive Decay with 1000 random numbers
import numpy as np
import matplotlib.pyplot as plt

tau = 3.053*60
N=1000
P = np.log(2)/tau
N_Tl=1000
tmax=1000

Distribution = []
Tl = N_Tl * np.ones([tmax-1])
Pb = np.zeros([tmax-1])

for i in range(N_Tl):
    Random = np.random.random()
    random_numbers = int(-np.log(1-Random)/P)
    Distribution.append(random_numbers)
    if random_numbers < tmax:
        Tl[random_numbers:] -= 1
        Pb[random_numbers:] += 1
        
        
        
plt.plot(Tl, label='Tl' , color='red')
plt.plot(Pb, label='Pb', color='blue')
plt.title('Decay of Tl & Pb')
plt.xlabel("Time (s)")
plt.ylabel("Number of atoms")

plt.legend()
plt.show()

plt.plot(Distribution)
plt.title('The Distribution of Random Numbers')
plt.xlim(0,1000)
plt.ylim(0,1000)
plt.show()
```
