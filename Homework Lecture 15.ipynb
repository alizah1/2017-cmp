```python
# Houman Alizadeh
# Homework Lecture 15
# 2D Function minimum using Monte Carlo and Gradient Decent Method
from random import random
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def Function(x,y):
    return (1/2)*x**3 + (1/4)*y**4  # just chose a random function

P = []

def MC(N=100000):       
    for i in range(N): 
        x = 2 * random()
        y = 2 * random()
        P.append(Function(x,y))
    return min(P)

print("The Minimum value of the function using Monte Carlo Method is:", MC())

x_min, x_max = -2, 2                        
y_min, y_max = -2, 2

nx = np.linspace(x_min,x_max,100)
ny = np.linspace(y_min,y_max,100)

x,y = np.meshgrid(nx,ny)

z = Function(x, y)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,z, cmap='summer', alpha=0.8)
cset = ax.contour(x,y,z, zdir='z', offset=0)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f')
plt.show()
```
