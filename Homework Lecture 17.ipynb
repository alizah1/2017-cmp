```python
import numpy as np
import matplotlib.pyplot as plt
L = 5
def LJ(r):
    r6 = r**6
    r12 = r6*r6
    return 4*(1/r12 - 1/r6) # since sigma = epsilon = 1

r = np.linspace(1.0,3.0,100)
# Now we have to minimize the function:
values = []
def Potential_Energy(N): # N = number of atmos
    for i in range(N):
        atom_1 = L*np.random.random_sample((3))
        atom_2 = L*np.random.random_sample((3))
        atom_3 = L*np.random.random_sample((3))
        r_1 = np.linalg.norm(atom_1-atom_2) # To find the distance between atom 1 & 2
        r_2 = np.linalg.norm(atom_3-atom_2) # To find the distance between atom 2 & 3
        r_3 = np.linalg.norm(atom_3-atom_1) # To find the distance between atom 1 & 3
        Energy_1 = LJ(r_1)
        Energy_2 = LJ(r_2)
        Energy_3 = LJ(r_3)
        Total_Energy = Energy_1 + Energy_2 + Energy_3
        values.append(Total_Energy)
    return min(values)
plt.plot(r, LJ(r))
plt.ylabel("Potential Energy")
plt.xlabel("Interatomic Separation")
plt.title("Lennard-Jones Potential Graph, $\epsilon$ = 1, $\delta$ = 1")
plt.show()

Potential_Energy(100000)
```
