```python
# Houman Alizadeh
# Homework Lecuture 8
# This is the linear plot fitting for a simple kinematic problem.
import numpy as np
import matplotlib.pyplot as plt

Position = lambda t: (3/2)*t**2+10*t+5
start, end = 0, 15
data_points = 10
t = np.linspace(start , end , data_points)
X = Position(t) + np.random.rand(data_points)+10
fit = np.polyfit(t,X,2)
X_exp = np.polyval(fit,t)


plt.plot (t , Position(t), "-", label = "Linear Regression")
plt.plot (t, X, "o", label = "Experimental data")
#plt.plot (t, X_exp, 'o', label = "Experimental data")

plt.xlabel("time [s]")
plt.ylabel("Position [m]")
plt.title("Position vs. Time")
plt.legend()
plt.show()
```

```python
Velocity = lambda t: 3*t+10
start, end = 0,15
data_points = 10
t = np.linspace(start , end , data_points)
V = Velocity(t) + np.random.rand(data_points)+1.0
fit = np.polyfit(t,V,4)
V_exp = np.polyval(fit,t)


plt.plot (t , Velocity(t), "-", label = "Linear Regression")
#plt.plot (t, V, "o", label = "Experiment data")
plt.plot (t, V_exp, 'o', label = "Experimental data")

plt.xlabel("time [s]")
plt.ylabel("Velocity [m/s]")
plt.title("Velocity vs. Time")
plt.legend()
plt.show()
```
